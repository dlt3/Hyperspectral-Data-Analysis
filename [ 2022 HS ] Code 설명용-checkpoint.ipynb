## **모델**

### **_기본 설정**

```python
### 기본 설정
Random_state = 717      # 같은 결과를 유지하기 위해 Seed 고정.
lag_term = 3            # lag data 생성에 사용할 시점 결정 (1이상의 정수).
before_y = "no"         # lag data 생성 시 이전 목적 변수를 사용할지 여부 결정 (yes or no).
hs = "yes"              # 변수 중 Hyperspectral 변수를 사용할지 여부 결정 (yes or no).
deep_learning = "no"   # 딥러닝 사용할지 여부 설정 (yes or no) - yes 시 머신러닝은 사용하지 않음.

### 훈련에 사용할 훈련 모델 설정 - 여러개 설정 시, 앙상블(ensemble) 수행
MODELS = [RandomForestRegressor(random_state = Random_state), xgboost.XGBRegressor(random_state = Random_state), LGBMRegressor(random_state = Random_state)]

# MODELS = [RandomForestRegressor(random_state = Random_state), ExtraTreesRegressor(random_state = Random_state), AdaBoostRegressor(random_state = Random_state), xgboost.XGBRegressor(random_state = Random_state), 
#           GradientBoostingRegressor(random_state = Random_state), LGBMRegressor(random_state = Random_state), CatBoostRegressor(random_state = Random_state, verbose = 0)]
```

### **_기본 모델**
- 테스트 데이터인 4월 데이터 중 첫날 데이터를 예측하기 위한 기본 훈련된 모델 설정
- 3월 데이터로 기본 모델 생성.

```python
### 기본 모델
model = trc.HS_model(data_yp_train,                                                         # 기본 모델 생성에 사용할 데이터 (data_yp_train : 3월 데이터)
                     MODEL = GradientBoostingRegressor(random_state = Random_state),        # 기본 모델로 사용할 모델 : Gradient Boosting Regressor
                     lag_term = lag_term,                                                   # 모델 훈련에 사용할 설명 변수의 lag term 설정 (lag_term = 3)
                     before_y = before_y,                                                   # 모델 훈련에 사용할 설명 변수의 이전 타겟변수 사용 여부 설정 (before_y = "no")
                     hs = hs,                                                               # 모델 훈련에 사용할 설명 변수의 Hyperspectral 변수 사용 여부 설정 (hs = "yes")
                     target = target,                                                       # 모델 훈련에 사용할 목적 변수 설정 (target = "chlorophyll")
                     deep_learning = deep_learning)["model"]                                # 딥러닝 사용할지 여부 설정 (deep_learning = "yes")            
print(model)
```

### **_모델 훈련 및 예측 과정**

- 매일 자정(0시 0분) 모델 업데이트.
- 매번 15분 후 목적 변수를 예측.

```python
### 경고문 출력 방지
import warnings
warnings.filterwarnings('ignore')




### 예측 과정
dataset = data_yp_train.copy()  # 데이터가 쌓일 공간
preds = []                      # 예측값이 저장될 공간
pred_X_df = pd.DataFrame()      # 예측에 사용된 설명변수가 저장될 공간




### 반복문을 통해 4월 데이터(data_yp_test)가 하나씩 들어온다고 가정.
for i in tqdm(range(data_yp_test.shape[0])) :

    
    dataset = dataset.append(data_yp_test.iloc[[i]])   # 들어오는 데이터를 저장.
    last_data = dataset.iloc[[-1]]                     # 가장 마지막에 들어온 데이터를 따로 저장.
    
    
    """ 매 0시0분 Model Update """
    if (last_data["Hour"].values == 0) & (last_data["Min"].values == 0) :
        print("\033[31m" +"%d년 %d월 %d일 %d시 %d분 Model Update 수행합니다." % (last_data["Year"], last_data["Month"], last_data["Day"], last_data["Hour"], last_data["Min"]) + "\033[0m")
        
        start_time = time.time()  # 모델 update가 시작되는 시간
        
        # 모델 훈련 업데이트 (모델을 여러개 설정 시 반복문을 통해 여러 모델로 훈련)
        hs_models, models = [], []
        if deep_learning == "yes" :
            hs_model = trc.HS_model(dataset, MODEL = MODELS[j], lag_term = lag_term, before_y = before_y, hs = hs, target = target, deep_learning = deep_learning)
            hs_models.append(hs_model)
            models.append(hs_model["model"])
        
        else : 
            for j in range(len(MODELS)) :
                hs_model = trc.HS_model(dataset, MODEL = MODELS[j], lag_term = lag_term, before_y = before_y, hs = hs, target = target, deep_learning = deep_learning)
                hs_models.append(hs_model)
                models.append(hs_model["model"])

        # 예측에 사용되는 데이터를 변경하기 위해 훈련 데이터 정보 저장
        train_pls = hs_models[0]["train pls"]
        train_std_other = hs_models[0]["train std other"]
        train_std_reflec = hs_models[0]["train std reflec"]
        train_columns = hs_models[0]["train columns"]
        
        end_time = time.time()   # 모델 update가 끝나는 시간
        print("%.3f 경과시간" % (np.round(end_time - start_time, 3)))   # 모델 update에 걸린 시간

        
    else :
        pass
    
    
    
    """ Prediction """
    # 가장 마지막에 들어온 데이터를 훈련 데이터 정보를 사용해 변수 세팅 후 15분 후의 목적 변수 예측
    pred_X = tec.testlag_(dataset, last_data, lag_term = lag_term, before_y = before_y, target = target)     # lag data로 변경
    pred_X = tec.prep_data_with_y(pred_X, train_pls, train_std_other, train_std_reflec, hs = hs)             # lag data에 훈련 정보를 사용해 변수 세팅하여 예측용 변수 생성.
    pred_X = pred_X[train_columns]                                                                           # 예측용 변수를 훈련용 변수와 같은 순서로 재배열
    pred_X_df = pd.concat([pred_X_df, pred_X], axis = 0)                                                     # 예측용 변수들을 저장해둘 공간 (추후 문제 발생 시 데이터 확인을 위해 저장)
    
    if deep_learning == "yes" :                                                                               # 머신러닝이 아닌 딥러닝 사용 시 예측용 변수들 차원(shape) 변경
        pred_X = pred_X.values.reshape((pred_X.shape[0], pred_X.shape[1], 1))
        
    
    pred_list = []                                                # (여러 모델의) 예측값이 저장될 공간
    for j in range(len(models)) :                                 # 모델을 여러개 설정시 각 모델에 대하여 예측값 생성 후 저장
        pred_ = models[j].predict(pred_X)[0]
        pred_list.append(pred_)
        
    pred = np.mean(pred_list)                                      # 여러 모델로 예측된 여러 예측값의 평균을 최종 에측값으로 사용.
    
    preds.append(pred)                                             # 최종 예측값을 저장.
    
    
    # 50번째 데이터마다 현재 상황 출력
    if i%50 == 0 :                             
        print("%d년 %d월 %d일 %d시 %d분 + 15분 후 예측값 : %.4f" % (last_data["Year"], last_data["Month"], last_data["Day"], last_data["Hour"], last_data["Min"], pred))
```

## **예측 결과**

- 4월 데이터 중 첫번째 데이터를 이용하여 예측 시작 즉, 4월 데이터 중 두번째 데이터부터 예측

- 4월 데이터 중 마지막 데이터를 사용해 보유한 데이터보다 하나 더 예측

```python
# 실제 데이터 - 4월 데이터 중 두번째 데이터부터 예측되었으므로 비교를 위해 실제 데이터의 첫번째 값은 제외.
observed = list(data_yp_test.loc[1:, target].reset_index(drop = True))

# 예측 값 - 4월 마지막 데이터를 통해 실제 데이터보다 한 시점 더 예측하였으므로 비교를 위해 예측값의 마지막은 제외.
preds2 = preds[:-1]
```

```python
# 예측 평가 - R-squre, Correlation, MAPE
print("R-square(OLS) : ", mc.R2_OLS(observed, preds2))
print("R-square(sklearn) : ", np.round(r2_score(observed, preds2), 3))
print("Correlation : ", np.round(np.corrcoef(observed, preds2)[0,1], 3))
print("MAPE : ", mc.MAPE(observed, preds2))   # 실제 값 중 마지막값이 0이기에 이 부분때문에 값이 크게 나오는것으로 확인

# 실제 값 중 마지막값이 0이기에 이 부분때문에 값이 크게 나오는것으로 확인
# 마지막 값 제외
print("\033[31m" + "================ 마지막 값 제외 ================" + "\033[0m")
print("R-square(OLS) : ", mc.R2_OLS(observed[:-1], preds2[:-1]))
print("R-square(sklearn) : ", np.round(r2_score(observed[:-1], preds2[:-1]), 3))
print("Correlation : ", np.round(np.corrcoef(observed[:-1], preds2[:-1])[0,1], 3))
print("MAPE : ", mc.MAPE(observed[:-1], preds2[:-1]))


# 예측값과 실제값의 scatter plot
plt.figure(figsize = (5,5))
plt.scatter(observed, preds2)
plt.xlabel("Observed") ; plt.ylabel("predicted")
plt.plot([0, max(max(observed), max(preds2))], [0, max(max(observed), max(preds2))]);
```
